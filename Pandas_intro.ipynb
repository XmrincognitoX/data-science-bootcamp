```python
import pandas as pd
```

```python
import re
```

```python
dane = pd.DataFrame(data=[[10, 4, 8], [12, 5, 7]], index=['first', 'second'], columns=['a  t', 'b', 'c'])
dane
```

```python
dane.columns = [col if col!='a' else 'ab' for col in dane.columns]
dane
```

```python
dane.columns = [col.replace(' ','_') for col in dane.columns]
dane
```

```python
dane.columns = [re.sub('\s+','_', col) for col in dane.columns]
dane
```

```python
lista=[]
for col in dane.columns:
  print(col)
  print(lista)
  if col=='a':
    col='ab'
  lista.append(col)
  print(lista)
dane.columns = lista
dane
```
